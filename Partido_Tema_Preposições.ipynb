```python
import pandas as pd

df = pd.read_excel('proposicoes-2023.xlsx')
df.head(5)
```

```python
cruzamento = pd.crosstab(df['tema'], df['partido'])
cruzamento.loc['Total'] = cruzamento.sum()
cruzamento['Soma da Linha'] = cruzamento.sum(axis=1)
cruzamento
```

```python
cruzamento_filtrado_coluna = cruzamento.loc[:, cruzamento.loc['Total'] > 200]

cruzamento_filtrado = cruzamento_filtrado_coluna.loc[cruzamento_filtrado_coluna['Soma da Linha'] > 300]

cruzamento_filtrado
```

```python
cruzamento_filtrado = cruzamento_filtrado.loc[cruzamento_filtrado.index != 'Total']
cruzamento_filtrado = cruzamento_filtrado.drop('Soma da Linha', axis=1)

cruzamento_filtrado
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Gerar mapa de calor
sns_plot = sns.heatmap(cruzamento_filtrado , vmin=20, vmax=500,annot=True, cmap="YlGnBu", annot_kws={"size": 8},fmt='.0f')

plt.title('Temas (proposições) Vs Partidos', fontsize=14, pad=20)

sns_plot.set_xlabel('')
sns_plot.set_ylabel('')

sns_plot.set_xticklabels(sns_plot.get_xticklabels(), size=8, rotation=45)
sns_plot.set_yticklabels(sns_plot.get_yticklabels(), size=8)

# Configurar a fonte da escala de cores (legenda)
cbar = sns_plot.collections[0].colorbar
cbar.ax.tick_params(labelsize=8)


plt.show()
```
